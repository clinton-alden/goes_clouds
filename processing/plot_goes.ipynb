```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import geogif
```

## Open file

```python
file_path = '/home/cdalden/goes_work/goes16/2023/6/10/ABI-L1b-RadC/00/C13/'
file_name = 'OR_ABI-L1b-RadC-M6C13_G16_s20231610001184_e20231610003569_c20231610004020.nc'
ds = xr.open_dataset(file_path + file_name)
ds
```

```python
# Extract the x and y coordinates and the variable
x = ds.variables['x'][:]
y = ds.variables['y'][:]
var = ds.variables['Rad'][:]

# Create a meshgrid for plotting
X, Y = np.meshgrid(x, y)

# Plot the variable
plt.figure(figsize=(10, 6))
plt.contourf(X, Y, var, cmap='viridis')
plt.colorbar(label='Variable Units')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Plot of Variable')
plt.show()
```
